# Windowing of the signals

```python
def find_n_windows(l_signal, w_window=3600, Maxoverlap=0.5):
    n_windows = np.ceil(l_signal / w_window)
    if n_windows ==1:
        overlap = 0
    else:
        overlap = np.ceil(((n_windows * w_window) - l_signal) / (n_windows -1))
    if overlap * (n_windows - 1) > (w_window * Maxoverlap):
        n_windows = n_windows -1
        overlap = 0
    return n_windows, overlap


#import numpy as np

def WindowSignals(X, y, Maxoverlap=0.5):
    
    if isinstance(X, pd.DataFrame) or isinstance(X, pd.Series):
        X = X.to_numpy()  # Convert to numpy array if X is a DataFrame or Series
    if isinstance(y, pd.DataFrame) or isinstance(y, pd.Series):
        y = y.to_numpy()  # Convert to numpy array if y is a DataFrame or Series
    

    X_post_win = []
    y_post_win = []

    for i in range(len(X)):  
        signal = X[i]
        label = y[i]  

        l_signal = len(signal)  
        w_window = 3600 

        # Calculates the number of windows and overlap based on the signal length and window size
        if Maxoverlap > 0:
            n_windows, overlap = find_n_windows(l_signal, w_window, Maxoverlap)
        else:
            n_windows = np.floor(l_signal / w_window)
            overlap = 0

        # Generate the windows
        for n in range(int(n_windows)):
            start_index = int(n * (w_window - overlap))
            end_index = int(start_index + w_window)
            if end_index > l_signal:  
                break
            window = signal[start_index:end_index]
            X_post_win.append(window)
            y_post_win.append(label)  # Appends the same label for all windows of the same signal

    return np.array(X_post_win, dtype=object), np.array(y_post_win, dtype=object)
```

```python
# Windowing of the signals of the datasets
X_windows_train, y_windows_train= WindowSignals(train['signals'], train['y'])
X_windows_validation, y_windows_validation = WindowSignals(val['signals'], val['y'])
X_windows_test, y_windows_test = WindowSignals(test['signals'], test['y'])
X_windows_test_sick, y_windows_test_sick = WindowSignals(test_sick['signals'], test_sick['y'])
```

# Data generator

```python
import tensorflow as tf
# Define the noise-adding function
def AddLinearInterp(X, n_segments, len_segments):
    """
    Adds linear interpolation noise to the data.
    """
    X_noisy = X.copy()  # Work on a copy to avoid modifying the original data
    for i in range(n_segments):
        start = np.random.randint(0, X_noisy.shape[0] - len_segments)
        stop = start + np.random.randint(1, len_segments)
        X_noisy[start:stop + 1] = np.linspace(X_noisy[start], X_noisy[stop], stop - start + 1)
    return X_noisy
 
# Define the data generator class
class NoisyDataGenerator(tf.keras.utils.Sequence):
    def __init__(self, X, y, batch_size, add_noise_func = AddLinearInterp, n_segments=10, len_segments=120, shuffle=True, addNoise = True):
        """
        Initializes the generator.
        Parameters:
        - X: Input data.
        - y: Labels.
        - batch_size: Number of samples per batch.
        - add_noise_func: Function to add noise to the data.
        - n_segments: Number of noisy segments to add.
        - len_segments: Maximum length of each noisy segment.
        - shuffle: Whether to shuffle the data at the beginning of each epoch.
        """
        self.X = np.asarray(X)
        self.y = np.asarray(y)
        self.batch_size = batch_size
        self.add_noise_func = add_noise_func
        self.n_segments = n_segments
        self.len_segments = len_segments
        self.shuffle = shuffle
        self.indices = np.arange(len(self.X))
        self.addNoise = addNoise
 
        if self.shuffle:
            np.random.shuffle(self.indices)
 
    def __len__(self):
        """Returns the number of batches per epoch."""
        return int(np.ceil(len(self.X) / self.batch_size))
 
    def __getitem__(self, index):
        """Generates one batch of data."""
        start_idx = index * self.batch_size
        end_idx = min(start_idx + self.batch_size, len(self.X))
        batch_indices = self.indices[start_idx:end_idx]
        X_batch = self.X[batch_indices]
        y_batch = self.y[batch_indices]
        if self.addNoise:
            # Add noise to each sample in the batch
            X_batch_noisy = np.array([self.add_noise_func(x, self.n_segments, self.len_segments) for x in X_batch])
            return X_batch_noisy, y_batch
        else:
            return X_batch, y_batch  

 
    def on_epoch_end(self):
        """Shuffles the data at the end of each epoch."""
        if self.shuffle:
            np.random.shuffle(self.indices)
```

# Model

The model is an Inception ResNet

```python
import tensorflow as tf

class RandomLinearInterpolationNoise(tf.keras.layers.Layer):
    def __init__(self, Interp_probability=0.1, Max_inter_len=3*2*60, **kwargs):
        super(RandomLinearInterpolationNoise, self).__init__(**kwargs)
        self.Interp_probability = Interp_probability
        self.Max_inter_len = Max_inter_len

    def call(self, inputs, training=None):
        if not training:
            return inputs
        else:
            if tf.random.uniform(shape=[1], minval=0., maxval=1.) < self.Interp_probability:
                shape = tf.shape(inputs)
                batch_size, num_points, num_channels = shape[0], shape[1], shape[2]
                
                inter_len = tf.random.uniform([], minval=1, maxval=self.Max_inter_len, dtype=tf.int32)
                MaxVal = num_points - inter_len
                start_point = tf.random.uniform([], minval=1, maxval=MaxVal, dtype=tf.int32)
                end_point = start_point + inter_len

                mask = tf.concat([
                    tf.zeros((batch_size, start_point, num_channels), dtype=tf.float32),
                    tf.ones((batch_size, inter_len, num_channels), dtype=tf.float32),
                    tf.zeros((batch_size, num_points - end_point, num_channels), dtype=tf.float32)
                ], axis=1)

                linearly_spaced_values_t = tf.linspace(inputs[:, start_point, :], inputs[:, end_point, :], inter_len, axis=1)
                linearly_spaced_values = tf.concat([
                    tf.zeros((batch_size, start_point, num_channels), dtype=tf.float32),
                    linearly_spaced_values_t,
                    tf.zeros((batch_size, num_points - end_point, num_channels), dtype=tf.float32)
                ], axis=1)

                return inputs * (1 - mask) + mask * linearly_spaced_values
            else:
                return inputs


def Conv_1D_Block(x, filters, kernel, strides=1, padding="same", dropout_rate=False):
    x = tf.keras.layers.Conv1D(filters, kernel, strides=strides, padding=padding, kernel_initializer="he_normal", kernel_regularizer=tf.keras.regularizers.l2(0.001))(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Activation('relu')(x)
    if dropout_rate:
        x = tf.keras.layers.Dropout(dropout_rate)(x)
    return x

class Inception_ResNet:
    def __init__(self, length, num_channel, n_filters, problem_type='Regression', output_nums=1, pooling='avg', dropout_rate=False):
        self.length = length
        self.num_channel = num_channel
        self.n_filters = n_filters
        self.problem_type = problem_type
        self.output_nums = output_nums
        self.pooling = pooling
        self.dropout_rate = dropout_rate

    def MLP(self, x):
        x = tf.keras.layers.GlobalAveragePooling1D()(x) if self.pooling == 'avg' else tf.keras.layers.GlobalMaxPooling1D()(x)
        x = tf.keras.layers.Flatten()(x)
        if self.dropout_rate:
            x = tf.keras.layers.Dropout(self.dropout_rate)(x)
        outputs = tf.keras.layers.Dense(self.output_nums, activation='softmax' if self.problem_type == 'Classification' else 'linear')(x)
        return outputs

    def Inception_ResNet_v1(self):
        inputs = tf.keras.Input((self.length, self.num_channel))
        inputs = RandomLinearInterpolationNoise()(inputs)
        x = Conv_1D_Block(inputs, self.n_filters * 1, 3, strides=2, padding='valid')
        x = Conv_1D_Block(x, self.n_filters * 1, 3, padding='valid')
        x = Conv_1D_Block(x, self.n_filters * 2, 3)
        x = tf.keras.layers.MaxPooling1D(3, strides=2)(x)
        x = Conv_1D_Block(x, self.n_filters * 2.5, 1)
        x = Conv_1D_Block(x, self.n_filters * 6, 3, padding='valid')
        x = Conv_1D_Block(x, self.n_filters * 8, 3, strides=2, padding='valid')

        for _ in range(2):
            x = Conv_1D_Block(x, self.n_filters * 2, 1)
            x = Conv_1D_Block(x, self.n_filters * 2, 3)
            x = Conv_1D_Block(x, self.n_filters * 2, 3)

        x = tf.keras.layers.MaxPooling1D(3, strides=2)(x)
        for _ in range(4):
            x = Conv_1D_Block(x, self.n_filters * 4, 1)
            x = Conv_1D_Block(x, self.n_filters * 4, 7)

        x = tf.keras.layers.MaxPooling1D(3, strides=2)(x)
        for _ in range(2):
            x = Conv_1D_Block(x, self.n_filters * 6, 1)
            x = Conv_1D_Block(x, self.n_filters * 6, 3)

        final_output = self.MLP(x)
        return tf.keras.Model(inputs, final_output, name='Inception_ResNet_v1')
```

# Training

```python
X_windows_train = tf.convert_to_tensor(X_windows_train, dtype=tf.float32)
X_windows_validation = tf.convert_to_tensor(X_windows_validation, dtype=tf.float32)
X_windows_test = tf.convert_to_tensor(X_windows_test, dtype=tf.float32)
X_windows_test_sick = tf.convert_to_tensor(X_windows_test_sick, dtype=tf.float32)
```

```python
model = model.Inception_ResNet_v1()
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.callbacks import ReduceLROnPlateau
import keras.backend as K

history = {}
train_generator = NoisyDataGenerator(
    X_windows_train, 
    y_windows_train, 
    batch_size=32, 
    n_segments=10, 
    len_segments=120, 
    shuffle=True,
    addNoise=True
)
val_generator = NoisyDataGenerator(
    X_windows_validation, 
    y_windows_validation, 
    batch_size=32, 
    n_segments=10, 
    len_segments=120, 
    shuffle=False,
    addNoise=False
)

reduce_lr = ReduceLROnPlateau(monitor='val_loss',  
                              factor=0.5,         
                              patience=5,          
                              min_lr=1e-6,         
                              verbose=1)     




model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0006103052350290162),
    loss='mse', 
    metrics=['mae']
)
early_stopping = EarlyStopping(
    monitor='val_mae',       
    patience=15,              
    restore_best_weights=True  
)
new_history = model.fit(train_generator, epochs=100, batch_size=32, validation_data=val_generator,  callbacks=[early_stopping, reduce_lr]).history
history.update({k: history.get(k, []) + new_history.get(k, []) for k in set(history) | set(new_history)})
```

```python
import scipy.stats as stats

def calculate_p_value(R_squared, n, k):
    
    F_stat = (R_squared / k) / ((1 - R_squared) / (n - k - 1))
    
    p_value = 1 - stats.f.cdf(F_stat, dfn=k, dfd=n-k-1)
    
    return p_value

R_squared = r2
n = len(true_labels)  
k = 1 

p_value = calculate_p_value(R_squared, n, k)
print(f"P-value: {p_value}")

R_squared_sick = r2_sick
n_sick = len(true_labels_sick) 
k = 1   

p_value_sick = calculate_p_value(R_squared_sick, n_sick, k)
print(f"P-value: {p_value_sick}")
```
