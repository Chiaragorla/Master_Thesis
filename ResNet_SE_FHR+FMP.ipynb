Signals are already interpolated and Normalized

```python
def find_n_windows(l_signal, w_window=3600, Maxoverlap=0.5):
    n_windows = np.ceil(l_signal / w_window)
    if n_windows ==1:
        overlap = 0
    else:
        overlap = np.ceil(((n_windows * w_window) - l_signal) / (n_windows -1))
    if overlap * (n_windows - 1) > (w_window * Maxoverlap):
        n_windows = n_windows -1
        overlap = 0
    return n_windows, overlap


#import numpy as np

def WindowSignals(X, y, Maxoverlap=0.5):
    
    if isinstance(X, pd.DataFrame) or isinstance(X, pd.Series):
        X = X.to_numpy()  # Convert to numpy array if X is a DataFrame or Series
    if isinstance(y, pd.DataFrame) or isinstance(y, pd.Series):
        y = y.to_numpy()  # Convert to numpy array if y is a DataFrame or Series
    

    X_post_win = []
    y_post_win = []

    for i in range(len(X)):  
        signal = X[i]
        label = y[i]  
        signal = np.array(signal)
        l_signal = signal.shape[0]  
        w_window = 3600 

        # Calculates the number of windows and overlap based on the signal length and window size
        if Maxoverlap > 0:
            n_windows, overlap = find_n_windows(l_signal, w_window, Maxoverlap)
        else:
            n_windows = np.floor(l_signal / w_window)
            overlap = 0

        # Generate the windows
        for n in range(int(n_windows)):
            start_index = int(n * (w_window - overlap))
            end_index = int(start_index + w_window)
            if end_index > l_signal:  
                break
            window = signal[start_index:end_index]
            X_post_win.append(window)
            y_post_win.append(label)  # Appends the same label for all windows of the same signal

    return np.array(X_post_win, dtype=object), np.array(y_post_win, dtype=object)
```

```python
# Windowing on the signals of the datasets
X_windows_train_FHR, y_windows_train= WindowSignals(train['FHR'], train['y'])
X_windows_validation_FHR, y_windows_validation = WindowSignals(val['FHR'], val['y'])
X_windows_test_FHR, y_windows_test = WindowSignals(test['FHR'], test['y'])
```

# Model

The model is a ResNet with SE attention blocks

```python
X_train = np.stack([X_windows_train_FHR, X_windows_train_FMP], axis=-1)  # (batch_size, timesteps, 2)
X_val = np.stack([X_windows_validation_FHR, X_windows_validation_FMP], axis=-1)  # (batch_size, timesteps, 2)
X_test = np.stack([X_windows_test_FHR, X_windows_test_FMP], axis=-1)  # (batch_size, timesteps, 2)
```

```python
import tensorflow as tf
class RandomLinearInterpolationNoise(tf.keras.layers.Layer):
    def __init__(self, Interp_probability=0.1, Max_inter_len=3*2*60, **kwargs):
        super(RandomLinearInterpolationNoise, self).__init__(**kwargs)
        self.Interp_probability = Interp_probability
        self.Max_inter_len = Max_inter_len

    def call(self, inputs, training=None):
        if not training:
            return inputs
        else:
            if tf.random.uniform(shape=[1], minval=0., maxval=1.) < self.Interp_probability:
                shape = tf.shape(inputs)
                batch_size, num_points, num_channels = shape[0], shape[1], shape[2]
                
                inter_len = tf.random.uniform([], minval=1, maxval=self.Max_inter_len, dtype=tf.int32)
                MaxVal = num_points - inter_len
                start_point = tf.random.uniform([], minval=1, maxval=MaxVal, dtype=tf.int32)
                end_point = start_point + inter_len

                mask = tf.concat([
                    tf.zeros((batch_size, start_point, num_channels), dtype=tf.float32),
                    tf.ones((batch_size, inter_len, num_channels), dtype=tf.float32),
                    tf.zeros((batch_size, num_points - end_point, num_channels), dtype=tf.float32)
                ], axis=1)

                linearly_spaced_values_t = tf.linspace(inputs[:, start_point, :], inputs[:, end_point, :], inter_len, axis=1)
                linearly_spaced_values = tf.concat([
                    tf.zeros((batch_size, start_point, num_channels), dtype=tf.float32),
                    linearly_spaced_values_t,
                    tf.zeros((batch_size, num_points - end_point, num_channels), dtype=tf.float32)
                ], axis=1)

                return inputs * (1 - mask) + mask * linearly_spaced_values
            else:
                return inputs


from tensorflow import keras

def se_block(input_tensor, reduction):
    filters = input_tensor.shape[-1]  
    se = keras.layers.GlobalAveragePooling1D()(input_tensor)  # Squeeze
    se = keras.layers.Dense(filters // reduction, activation='relu')(se)  # Bottleneck
    se = keras.layers.Dense(filters, activation='sigmoid')(se)  # Scaling
    se = keras.layers.Multiply()([input_tensor, keras.layers.Reshape((1, filters))(se)])  # Recalibration
    return se

#resnet
def build_model(input_shape, classes, drop_out_rate, Interp_probability=0.05, Max_inter_len=3*2*60, n_feature_maps = 32):
        
        input_layer = keras.layers.Input(input_shape)
        input_layer = keras.layers.GaussianNoise(0.05)(input_layer) #add Gaussian Noise
        input_layer = RandomLinearInterpolationNoise(Interp_probability=Interp_probability, Max_inter_len=Max_inter_len)(input_layer)
        # BLOCK 1

        conv_x = keras.layers.Conv1D(filters=n_feature_maps, kernel_size=8, padding='same')(input_layer)
        conv_x = keras.layers.BatchNormalization()(conv_x)
        conv_x = keras.layers.Activation('relu')(conv_x)

        conv_y = keras.layers.Conv1D(filters=n_feature_maps, kernel_size=5, padding='same')(conv_x)
        conv_y = keras.layers.BatchNormalization()(conv_y)
        conv_y = keras.layers.Activation('relu')(conv_y)

        conv_z = keras.layers.Conv1D(filters=n_feature_maps, kernel_size=3, padding='same')(conv_y)
        conv_z = keras.layers.BatchNormalization()(conv_z)

        # expand channels for the sum
        shortcut_y = keras.layers.Conv1D(filters=n_feature_maps, kernel_size=1, padding='same')(input_layer)
        shortcut_y = keras.layers.BatchNormalization()(shortcut_y)

        output_block_1 = keras.layers.add([shortcut_y, conv_z])
        output_block_1 = se_block(output_block_1, reduction=2*8)
        output_block_1 = keras.layers.Dropout(drop_out_rate)(output_block_1) #
        output_block_1 = keras.layers.Activation('relu')(output_block_1)

        # BLOCK 2

        conv_x = keras.layers.Conv1D(filters=n_feature_maps * 2, kernel_size=8, padding='same')(output_block_1)
        conv_x = keras.layers.BatchNormalization()(conv_x)
        conv_x = keras.layers.Activation('relu')(conv_x)

        conv_y = keras.layers.Conv1D(filters=n_feature_maps * 2, kernel_size=5, padding='same')(conv_x)
        conv_y = keras.layers.BatchNormalization()(conv_y)
        conv_y = keras.layers.Activation('relu')(conv_y)

        conv_z = keras.layers.Conv1D(filters=n_feature_maps * 2, kernel_size=3, padding='same')(conv_y)
        conv_z = keras.layers.BatchNormalization()(conv_z)

        # expand channels for the sum
        shortcut_y = keras.layers.Conv1D(filters=n_feature_maps * 2, kernel_size=1, padding='same')(output_block_1)
        shortcut_y = keras.layers.BatchNormalization()(shortcut_y)

        output_block_2 = keras.layers.add([shortcut_y, conv_z])
        output_block_2 = se_block(output_block_2, reduction=3*8)
        output_block_2 = keras.layers.Dropout(drop_out_rate)(output_block_2) #
        output_block_2 = keras.layers.Activation('relu')(output_block_2)

        # BLOCK 3

        conv_x = keras.layers.Conv1D(filters=n_feature_maps * 2, kernel_size=8, padding='same')(output_block_2)
        conv_x = keras.layers.BatchNormalization()(conv_x)
        conv_x = keras.layers.Activation('relu')(conv_x)

        conv_y = keras.layers.Conv1D(filters=n_feature_maps * 2, kernel_size=5, padding='same')(conv_x)
        conv_y = keras.layers.BatchNormalization()(conv_y)
        conv_y = keras.layers.Activation('relu')(conv_y)

        conv_z = keras.layers.Conv1D(filters=n_feature_maps * 2, kernel_size=3, padding='same')(conv_y)
        conv_z = keras.layers.BatchNormalization()(conv_z)

        # no need to expand channels because they are equal
        shortcut_y = keras.layers.BatchNormalization()(output_block_2)

        output_block_3 = keras.layers.add([shortcut_y, conv_z])
        output_block_3 = se_block(output_block_3, reduction=4*8)
        output_block_3 = keras.layers.Dropout(drop_out_rate)(output_block_3) #
        output_block_3 = keras.layers.Activation('relu')(output_block_3)

        # FINAL

        gap_layer = keras.layers.GlobalAveragePooling1D()(output_block_3)
        gap_layer = keras.layers.Dropout(drop_out_rate)(gap_layer) # To avoid overfitting
        output_layer = keras.layers.Dense(classes, activation='linear')(gap_layer)

        model = keras.models.Model(inputs=input_layer, outputs=output_layer)

        return model
```

# Training

```python
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.callbacks import ReduceLROnPlateau
import keras.backend as K
import gc
history = {}

reduce_lr = ReduceLROnPlateau(monitor='val_loss',  
                              factor=0.5,         
                              patience=10,          
                              min_lr=1e-6,         
                              verbose=1)     




model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0006103052350290162),
    loss='mse', 
    metrics=['mae']
)
early_stopping = EarlyStopping(
    monitor='val_loss',       
    patience=15,              
    restore_best_weights=True,
    mode='min'  
)
new_history = model.fit(X_windows_train, y_windows_train, epochs=200, batch_size=32, validation_data=(X_windows_validation, y_windows_validation),  callbacks=[early_stopping, reduce_lr]).history
history.update({k: history.get(k, []) + new_history.get(k, []) for k in set(history) | set(new_history)})
```

```python
import scipy.stats as stats

def calculate_p_value(R_squared, n, k):
    
    F_stat = (R_squared / k) / ((1 - R_squared) / (n - k - 1))
    
    
    p_value = 1 - stats.f.cdf(F_stat, dfn=k, dfd=n-k-1)
    
    return p_value

R_squared = r2
n = len(true_labels)  
k = 1  

p_value = calculate_p_value(R_squared, n, k)
print(f"P-value: {p_value}")

R_squared_sick = r2_sick
n_sick = len(true_labels_sick)  
k = 1    

p_value_sick = calculate_p_value(R_squared_sick, n_sick, k)
print(f"P-value: {p_value_sick}")
```
